## Cleaning

```python
# Convert Gender into numbers
df['Gender'] = df['Gender'].replace({'Male':0,'Female':1,'Other':2})
```

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)
```

```python
df.head()
```

## Univariate Analysis

```python
sns.histplot(df['Daily_Screen_Time(hrs)']) 
plt.show()
```

## Bivariate Analysis

```python
sns.scatterplot(x='Daily_Screen_Time(hrs)', y='Stress_Level(1-10)', data=df)
plt.show()
```

## Correlation Heatmap

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True)
plt.show()
```

## Correlation Conclusion (Pointwise with Numerical Values)

1. **More screen time leads to more stress**
   - Correlation: **+0.74**
   - A strong positive relationship indicating that higher screen time is linked with higher stress levels.

2. **More screen time leads to poorer sleep quality**
   - Correlation: **0.76**
   - A strong negative relationship showing that increased screen use is associated with worse sleep quality.

3. **More screen time leads to lower happiness**
   - Correlation: **0.71**
   - A strong negative relationship indicating that people with higher screen time tend to be less happy.

4. **Better sleep leads to higher happiness**
   - Correlation: **+0.68**
   - A strong positive relationship showing that people with good sleep quality usually report higher happiness.

5. **More stress leads to lower happiness**
   - Correlation: **0.74**
   - A strong negative relationship meaning that increased stress significantly reduces happiness.

6. **Better sleep is linked to lower stress**
   - Correlation: **0.58**
   - A moderate negative relationship indicating that better sleep quality is related to lower stress levels.

7. **Exercise frequency has very little impact**
   - Exercise and Happiness: **+0.041**
   - Exercise and Screen Time: **0.099**
   - These values show very weak relationships, meaning exercise does not strongly affect mental health indicators in this dataset.

8. **Age has almost no effect on mental health indicators**
   - Correlation values range from **0.05 to +0.06**
   - This shows that age does not play a meaningful role in stress, happiness, or sleep quality.

9. **Gender does not show any significant relationships**
   - Correlation values range from **0.05 to +0.03**
   - Gender does not have a noticeable impact on the mental health measures in this dataset.

10. **Days without social media do not show meaningful correlations**
    - Correlations range from **0.01 to +0.06**
    - This indicates that days away from social media do not strongly influence the mental health indicators in this dataset.


## Conclusion

The data shows that using screens too much can make people more stressed, sleep worse, and feel less happy. People who sleep well are usually happier. Stress reduces happiness a lot. Other things like age, gender, or exercise dont change much in this data. So, reducing screen time and improving sleep seems important for better mental health.
